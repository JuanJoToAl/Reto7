```python
# Se declaran variables e inicializa variable factor y producto

factor : int = 1
producto : int = 1
numero = int(input("Ingrese un nÃºmero"))

if numero >= 0:

    # Se ejecuta ciclo while hasta el factor sea mayor que el numero
    while numero >= factor:
        producto *= factor
        factor += 1
    print("El factorial de " + str(numero) + " es " +str(producto))

# Si se ingresa un nÃºmero negativo, se imprime mensaje
else:
    print("Los nÃºmero negativos no tienen factorial")
```
